```python
import cv2
```

```python
eye_feature  = cv2.CascadeClassifier('haarcascade_eye.xml')
front_feature = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
nose_feature = cv2.CascadeClassifier('haarcascade_mcs_nose.xml')
mouth_feature = cv2.CascadeClassifier('haarcascade_mcs_mouth.xml')
mask = cv2.CascadeClassifier('hand.xml')
```

# Detection Function        

```python
def detect(gray , frame):
    faces = front_feature.detectMultiScale(gray,1.7,5)
    for (x, y , w, h) in faces:
        cv2.rectangle(frame, (x,y),(x+w,y+h),(255,0,0),2)##first image second upper corner third lower fourth rgb
        roi_gray = gray[y:y+h,x:x+w]
        roi_color = frame[y:y+h,x:x+w]
        eyes = eye_feature.detectMultiScale(roi_gray,1.1,3)
#         mouth = mouth_feature.detectMultiScale(roi_gray,1.1,3)
        nose = nose_feature.detectMultiScale(roi_gray,1.1,3)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
            eyes_color = frame[ey:ey+eh,ex:ex+ew]
            cv2.putText(
            eyes_color, "Eye Detected", (int(ex+30), int(ey+30)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 1)
        for (nx,ny,nw,nh) in nose:    
            cv2.circle(roi_color,(nx+30,ny+10),30,(255,255,255),2)
#         for (mx,my,mw,mh) in mouth:
#             cv2.rectangle(roi_color,(mx,my),(mx+mw,my+mh),1.1,2)
        cv2.putText(
        frame, "Face Detected", (int(x), int(y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 1)
    hand = mask.detectMultiScale(gray,1.1,3)
    for (hx,hy,hw,hh) in hand:
        cv2.rectangle(frame, (hx,hy),(hx+hw,hy+hh),(255,255,0),2)
    return frame       
```

```python
video_capture = cv2.VideoCapture(0) 
```

```python
while True:
    _,frame = video_capture.read()##second element is the last frame ignoring last frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # We do some colour transformations.
    canvas = detect(gray, frame) # We get the output of our detect function.
    cv2.namedWindow('Video', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('Video', 1600, 1000)
    fps = video_capture.get(cv2.CAP_PROP_FPS);
    print(fps)
    cv2.imshow('Video', canvas) # We display the outputs.
    if cv2.waitKey(1) & 0xFF == ord('q'): # If we type on the keyboard:
        break # We stop the loop.
video_capture.release() # We turn the webcam off.
cv2.destroyAllWindows() # We destroy all the windows inside which the images were displayed        
```
